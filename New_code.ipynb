```python
x=data.drop(['Cuisines','Locality', 'Longitude', 'Latitude', 'Country Code', 'Address','Locality Verbose','Switch to order menu','Currency'], axis=1)
x.head()
```

```python
# Since New Delhi has most number of restaurants partnered with zomato we can analyse which top 10 Localities has most restaurants within New Delhi 
def city_stats(df, city_name):
    x1=df[(df.City == city_name)]
    x2=x1.Locality.value_counts().head(10)
    return x2
```

```python
def city_plot(df):
    df.plot(kind='bar')
    plt.ylabel('Number of Restaurants')
    plt.xlabel('Locality')
    plt.tight_layout()
    plt.show()
```

```python
city_plot(city_stats(data, 'Gurgaon'))
```

```python
#Logistic Regression model

#Preprocessing 

# x_test = x[x['Rating text'] == 'Not rated']

#Conversion of categorical variables to ordinal variables
#Target encoding

x['city_encoded'] = x.groupby("City")['Aggregate rating'].transform("mean")
x['restaurant_name_encoded'] = x.groupby("Restaurant Name")['Aggregate rating'].transform("mean")
x['Has Table booking'] = x['Has Table booking'].map({'Yes':1,'No':0})
x['Has Online delivery'] = x['Has Online delivery'].map({'Yes':1, 'No':0})
x['Is delivering now'] = x['Is delivering now'].map({'Yes':1, 'No':0})
x_temp_1 = pd.get_dummies(x['Rating color'], drop_first = True, prefix = 'rating_code_')
x_temp_2 = pd.get_dummies(x['Rating text'], drop_first = True, prefix = 'rating_text_')
x.drop(columns = ['Restaurant Name', 'City', 'Country'],inplace = True)
```

```python
x_temp_1
```

```python
x_temp_2
```

```python
processed_df = pd.concat([x, x_temp_1, x_temp_2], axis=1).drop(columns = ['Rating color', 'Rating text'])
processed_df.head()
```

```python
test = processed_df[processed_df['rating_text__Not rated'] == 1]
x_test = test.drop(columns = ['Aggregate rating'], axis = 1)
y_test = test['Aggregate rating']
train = processed_df[processed_df['rating_text__Not rated'] == 0]
x_train = train.drop(columns = ['Aggregate rating'], axis = 1)
y_train = train['Aggregate rating']
# x = processed_df.drop(columns = ['Aggregate rating'], axis =1)
# y = processed_df['Aggregate rating']

# from sklearn.model_selection import train_test_split
# x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.3)
```

```python
# test.head()
# Grid search cross validation
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
# grid={"C":np.logspace(-3,3,7), "penalty":["l2"]}# l1 lasso l2 ridge
logreg=LinearRegression()
# logreg_cv=GridSearchCV(logreg,grid,cv=10)
logreg.fit(x_train,y_train)

# print("tuned hpyerparameters :(best parameters) ",logreg_cv.best_params_)
# print("accuracy :",logreg_cv.best_score_)
```

```python
# from sklearn.linear_model import LogisticRegression
prediction = logreg.predict(x_test)
r2_score = logreg.score(x_test, y_test)
```
