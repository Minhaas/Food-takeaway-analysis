```python
df_data=pd.read_csv('./new_data/zomato.csv',encoding="latin-1")
df_country_data=pd.read_excel('./new_data/Country-Code.xlsx')
```

```python
df_data.head()
```

```python
df_country_data.head()
```

```python
#merged both the datasets since we are going to analyze the data based on the country name
df_Zomato=pd.merge(df_data,df_country_data,on="Country Code",how="left")
df_Zomato.head(5)
```

```python
'''To check which country has highest number of restaurants associated with zomato, given that India contains 
highest number of restaurants with zomato, let's carry out analysis on India'''

df_Zomato1=df_Zomato.groupby("Country").agg(Restaurant_Count=("Restaurant ID","count")).sort_values(by="Restaurant_Count",ascending=False)


```

```python
df_Zomato1.plot(kind="bar")
```

```python
data = df_Zomato[df_Zomato['Country']=="India"].reset_index(drop=True).copy()
nRow1, nCol1 = data.shape
print(f'Data has {nRow1} rows and {nCol1} columns.')
```

```python
# Checking if dataset contains any null
data.isna().sum()
```

```python
#checking no. of restaurants partnered with Zomato per city
data['City'].value_counts()
```

```python
# To check no. of restaurants which have online delivery
data.groupby(['Has Online delivery']).size().reset_index()
```
